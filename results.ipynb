```python
import matplotlib.pyplot as plt

ks = [1,2,5,10,20,30,50,100]
i=0
with open("result.csv") as f:
    for l in f:
        values = l[:-1].split(",")
        values,label = [int(n) for n in values[1:]],values[0]
        plt.plot(values,label=label)
        i+=1
        if i%2 == 0:
            plt.plot(ks,label="initial") 
            plt.legend()
            plt.show()
```
